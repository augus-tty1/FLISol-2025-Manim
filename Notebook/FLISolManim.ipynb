```python
%%manim -qm -v WARNING ex1

#exemplo 1
#cenas e mobjects
from manim import *

class ex1(Scene):
    def construct(self):

        circulo = Circle(color='#002BFF', fill_opacity=.5)
        estrela = Star(n=7,color=GREEN, fill_opacity=.5, density=3).next_to(circulo,RIGHT,buff=0.5)


        #ate aqui nada foi colocado em cena
        self.play(Create(circulo),Create(estrela),run_time=3)
        self.play(FadeOut(circulo))
        self.wait(1)
        self.remove(estrela)
        self.wait(1)
```

```python
%%manim -qm -v WARNING ex2

#exemplo 2
#propriedades podem ocorrer com o animate
#O metodo animate vai animar mudancas que sao aplicadas a um mobject
from manim import *

class ex2(Scene):
    def construct(self):

        circulo = Circle(color='#002BFF', fill_opacity=.5)
        estrela = Star(n=7,color=GREEN, fill_opacity=.5, density=3)

        #usando animate
        self.play(Create(estrela))
        self.play(estrela.animate.flip())
        self.play(estrela.animate.shift(RIGHT))
        self.play(estrela.animate.shift(2*LEFT))
        self.play(estrela.animate.shift(RIGHT))
        self.play(estrela.animate.rotate(PI/2).set_color('#002BFF'), run_time=2)
        self.play(estrela.animate.set_fill(opacity=0))
        self.wait(1)

        #usando o tranform
        self.play(Transform(estrela,circulo))
        self.play(estrela.animate.set_fill(opacity=1))#atencao metodo foi usado em estrela
        self.wait(1)

        self.play(FadeOut(estrela),run_time=3)
        self.wait(1)
```

```python
%%manim -qm -v WARNING ex3

#exemplo 3
#O metodo animate interpola o estado final e inicial do objeto
#Outros metodos podem ser mais apropriados
from manim import *

class ex3(Scene):
    def construct(self):

        quad1 = Square(color='#002BFF', fill_opacity=.8).shift(2*LEFT)
        quad2 = Square(color=GREEN, fill_opacity=.8).shift(2*RIGHT)

        self.play(quad1.animate.rotate(PI), Rotate(quad2,angle=PI),run_time=2)
        self.wait(1)
```

```python
%%manim -qm -v WARNING ex4


#exemplo 4
#metodos e transformacoes algumas observacoes
from manim import *

class ex4(Scene):
    def t(self):
        circulo = Circle(color='#002BFF', fill_opacity=.5)
        estrela = Star(n=7,color=GREEN,fill_opacity=.5,density=3).next_to(circulo,RIGHT,buff=.5)
        self.play(Transform(circulo,estrela))
        self.play(circulo.animate.set_color('#002BFF'))
        self.play(FadeOut(circulo))

    def rt(self):
        circulo = Circle(color='#002BFF', fill_opacity=.5)
        estrela = Star(n=7,color=GREEN,fill_opacity=.5,density=3).next_to(circulo,RIGHT,buff=.5)
        self.play(ReplacementTransform(circulo,estrela))
        self.play(estrela.animate.set_color('#002BFF'))
        self.play(FadeOut(estrela))

    def construct(self):
        self.t()
        self.wait(1)
        self.rt()
```

```python
%%manim -qm -v WARNING ex5

from manim import *

class ex5(Scene):
    def construct(self):
        euler = MathTex(r"e^x =").shift(4*LEFT)
        serie = MathTex(r"x^0 + x^1 + \frac{1}{2} x^2 + \frac{1}{6} x^3 + \cdots + \frac{1}{n!} x^n + \cdots").next_to(euler,RIGHT)
        somatorio = MathTex(r"\sum_{n=0}^\infty \frac{x^n}{n!}").next_to(euler,RIGHT)

        self.play(Write(euler), Write(serie))
        self.play(Transform(serie,somatorio))
        self.wait(1)
```
