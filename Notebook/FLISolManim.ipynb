```python
%%manim -qm -v WARNING ex2

#exemplo 2
#propriedades podem ocorrer com o animate
#O metodo animate vai animar mudancas que sao aplicadas a um mobject
from manim import *

class ex2(Scene):
    def construct(self):

        circulo = Circle(color='#002BFF', fill_opacity=.5)
        estrela = Star(n=7,color=GREEN, fill_opacity=.5, density=3)

        #usando animate
        self.play(Create(estrela))
        self.play(estrela.animate.flip())
        self.play(estrela.animate.shift(RIGHT))
        self.play(estrela.animate.shift(2*LEFT))
        self.play(estrela.animate.shift(RIGHT))
        self.play(estrela.animate.rotate(PI/2).set_color('#002BFF'), run_time=2)
        self.play(estrela.animate.set_fill(opacity=0))
        self.wait(1)

        #usando o tranform
        self.play(Transform(estrela,circulo))
        self.play(estrela.animate.set_fill(opacity=1))#atencao metodo foi usado em estrela
        self.wait(1)

        self.play(FadeOut(estrela),run_time=3)
        self.wait(1)
```

```python
%%manim -qm -v WARNING ex4


#exemplo 4
#metodos e transformacoes algumas observacoes
from manim import *

class ex4(Scene):
    def t(self):
        circulo = Circle(color='#002BFF', fill_opacity=.5)
        estrela = Star(n=7,color=GREEN,fill_opacity=.5,density=3).next_to(circulo,RIGHT,buff=.5)
        self.play(Transform(circulo,estrela))
        self.play(circulo.animate.set_color('#002BFF'))
        self.play(FadeOut(circulo))

    def rt(self):
        circulo = Circle(color='#002BFF', fill_opacity=.5)
        estrela = Star(n=7,color=GREEN,fill_opacity=.5,density=3).next_to(circulo,RIGHT,buff=.5)
        self.play(ReplacementTransform(circulo,estrela))
        self.play(estrela.animate.set_color('#002BFF'))
        self.play(FadeOut(estrela))

    def construct(self):
        self.t()
        self.wait(1)
        self.rt()
```

#**Agrupamentos**

Esse exemplo Ã© em particular interessante, alÃ©m de mostrar como funcionam os agrupamentos podemos declarar coordenadas (posiÃ§Ãµes) precisamente usando um vetor.

Um grupo pode ser visto como uma lista indexada ou seja grupo_pontos[0]=p1 neste caso, enquanto grupo_pontos[3]=l1;

---

```python
%%manim -qm -v WARNING agrupamento

# uma funcionalidade util pode ser o agrupamento para quando se quer alterar varios objetos

from manim import *

class agrupamento(Scene):
    def construct(self):
        p1 = Dot([-1,0,0])
        l1 = Text("p1").scale(.3).next_to(p1, DOWN)

        p2 = Dot([0,0,0], color=RED)
        l2 = Text("p2").scale(.3).next_to(p2, DOWN)

        p3 = Dot([1,0,0])
        l3 = Text("p3").scale(.3).next_to(p3,DOWN)

        p4 = Dot([2,0,0])
        l4 = Text("p4").scale(.3).next_to(p4,DOWN)

        grupo_pontos = VGroup(p1,p3,p4,l1,l3,l4)


        self.play(Create(p2), Create(l2), Create(grupo_pontos))
        self.play(grupo_pontos.animate.shift([0,3,0]))
        self.wait(1)
```

#**Usando texto e LaTEX**

```python
%%manim -qm -v WARNING ex5

from manim import *

class ex5(Scene):
    def construct(self):
        texto_comum = Text('OlÃ¡, mundo !')
        euler = MathTex(r"e^x =").shift(4*LEFT)
        serie = MathTex(r"x^0 + x^1 + \frac{1}{2} x^2 + \frac{1}{6} x^3 + \cdots + \frac{1}{n!} x^n + \cdots").next_to(euler,RIGHT)
        somatorio = MathTex(r"\sum_{n=0}^\infty \frac{x^n}{n!}").next_to(euler,RIGHT)

        self.play(AddTextLetterByLetter(texto_comum), run_time=3)
        self.play(FadeOut(texto_comum))

        self.play(Write(euler), Write(serie))
        self.play(Transform(serie,somatorio))
        self.wait(1)
```

```python
%%manim -qm -v WARNING ex6

# Exemplo 6
# Um exemplo de transformacao mais avancada e esse relacionado ao movimento
# Podemos construir uma funcao matematica interssante para os pontos e para um argumento de tempo
from manim import *

class ex6(Scene):
    def construct(self):

        elipse = Ellipse(width=5, height=4, color='#002BFF', stroke_width=20)

        # Grosseiramente, uma homotopia significa deformar continuamente algo
        def homotopy(x,y,z,t):
            if t<=.25:
                progress=t/.25
                return(x,y+progress*.2*np.sin(x),z)
            else:
                wave_progress=(t-.25)/.75
                return(x,y+.2*np.sin(x+10*wave_progress),z)

        self.play(Homotopy(homotopy,elipse ,rate_func=linear, run_time=3))

        # Argumento foram
        # homotopy : um mapeamento que muda os pontos (x,y,z) de acordo com t
        # Um mobject que tera seus pontos alterados
        # rate_func: controla como o tempo eh distribuido, eh mais avancado
        # run_time eh um float (numero quebrado)
```

```python
%%manim -qm -v WARNING mv1

#primeiro exemplo de camera se movimentando
from manim import *

class mv1(MovingCameraScene):
    def construct(self):

        circulo = Circle(color='#002BFF', fill_opacity=.5).shift(2*LEFT)
        estrela = Star(n=7,color=GREEN, fill_opacity=.5, density=3).shift(2*RIGHT)
        fusao = Star(n=7, color='#002BFF', fill_opacity=.5, density=1)
        g=VGroup(circulo,estrela)

        self.play(FadeIn(circulo), FadeIn(estrela))

        #salva o estado da camera para recuperar
        self.camera.frame.save_state()

        #camera vai focar no circulo
        #funcoes de movimentacao como move_to, next_to e aling_to resgatam a coordenada do centro do mobject

        self.play(self.camera.frame.animate.move_to(circulo).set(width=circulo.width*2))
        self.play(circulo.animate.flip(),run_time=.5)

        self.wait(1)

        #agora na estrela
        self.play(self.camera.frame.animate.move_to(estrela).set(width=estrela.width*2))
        self.play(Rotate(estrela,angle=2*PI),run_time=2)

        self.wait(1)

        #restaura
        self.play(Restore(self.camera.frame))

        #agora as coisas acontecem com todo o frame sendo filmado
        self.play(ReplacementTransform(g,fusao),run_time=3)
        self.play(Uncreate(fusao))
        self.wait(1)
```

```python
%%manim -qm -v WARNING mv2

# Nesse exemplo usamos uma animacao chamada MoveAlongPath que e um pouco mais avancada
# Teremos que utilizar um updater (chamado a cada frame)
# Mas queremos ilustrar o uso da camera

from manim import *

class mv2(MovingCameraScene):
    def construct(self):

        self.camera.frame.save_state()

        eixo = Axes(x_range=[-1,10] , y_range=[-1,5])
        grafico = eixo.plot(lambda x: np.cos(x), color='#002BFF', x_range=[0,3*PI])

        ponto = Dot(eixo.i2gp(grafico.t_min, grafico), color=GREEN)

        self.add(eixo, grafico, ponto)
        #self.play(self.camera.auto_zoom(ponto))
        self.play(self.camera.frame.animate.move_to(ponto))
        self.wait(1)

        # ao inves da funcao lambda poderiamos ter criado um metodo separado
        self.camera.frame.add_updater(lambda a: a.move_to(ponto))

        self.play(MoveAlongPath(ponto, grafico, rate_func=linear))

        self.play(Restore(self.camera.frame))

        self.wait(1)
```
